```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("global_student_migration.csv")
df["starting_salary_usd"] = pd.to_numeric(df["starting_salary_usd"], errors="coerce")

clean_salaries = []
for salary in df["starting_salary_usd"]:
    if salary and salary > 0:
        clean_salaries.append(salary)

plt.hist(clean_salaries, bins=40, color="lightblue")
plt.xlabel("Starting Salary")
plt.ylabel("Number of Students")
plt.title("Distribution of Starting Salaries")
plt.show()
```

```python
field_counts = df["destination_country"].value_counts()
print(field_counts.head(10))

plt.figure(figsize=(5, 3))
plt.bar(field_counts.index, field_counts.values, color="blue")
plt.ylim(470, 550)
plt.xlabel("Field of Study")
plt.ylabel("Number of Students")
plt.title(""Number of Students by Destination Country"")
plt.xticks(rotation=45)
plt.show()
```
